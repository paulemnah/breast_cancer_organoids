```python
from IPython import display

# Create subplots
#fig, axs = plt.subplots(len(cns_data), 1, sharex=True, figsize=(10, 4 * len(cns_data)))

# Iterate through each .cns file
for i, cnsobj in enumerate(cns_data):
    fig = cnvlib.do_scatter(cnarr = cnr_data[i], segments = cns_data[i], segment_color = "darkblue")
    
    # Set y-axis limits
    y_axis_lower_limit = -4
    y_axis_upper_limit = 4
    ax = fig.gca()
    ax.set_ylim(y_axis_lower_limit, y_axis_upper_limit)  # Set y-axis limits
    
    plt.savefig("data/cnvkit/plots/ipy_scatter/"+cnsobj.meta['sample_id']+"_scatter.png", dpi=600)
    display.display(fig)
```
